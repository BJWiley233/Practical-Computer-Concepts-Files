**1. [20 pts] Define each of the following machine learning terms:**
* **dataset:** an organized set or collection of data in R dimensions, in which there are a set of records (the observation) that each have N number of variable attributes
* **training dataset:** the random selection of datapoints/examples (also known as an observation) either with a target (supervised) or with out a target (unsupervised) used to train a machine learning classifier.
* **testing dataset:** a random selection of new datapoints/examples for which to test and predict the target using the classifier that we have trained.  Should be held away from the training and validation datasets until the final model evaluation.
* **validation dataset:**  A subset or subsets of the training dataset held back from the training datasets when training your classifier.  Then the validation subset is used to validate and fine tune the classifier until result is ready to be tested.
* **ground truth:**  The actual real true value of the target for a datapoint/example.
* **label:**  The value given for the target or the desired value for the attribute that we want to predict.
* **pre-processing:**  The processing of the dataset which is done before training the classifier.  This includes making sure the data fits the correct dimensions, feature engineering in the form of feature selection and dimensionality reduction or transformation, as well as check for missing data and noise.
* **feature:** An independent predictor variable (also known as attribute) from the dataset that is used to help predict the value for the target or response variable.  Each datapoint will have data with respect to the feature whether it value, classification, or NAN value.
* **numerical:**  A value that is in the set of $\mathbb{R}$.  Can be decimal, integer, radical, or fraction.
* **nominal:** Refers to data that is categorical and that cannot be ordered or sorted.  I believe this be assuming that alphabetical order does not count or gives no bias toward the target.  The text gives and example of “red” and “blue” which can be sorted alphabetically but this sorting would not give us any bias. 
* **decision surface:**  The boundary(ies) in multidimensions that separate the data into different regions.  These can be linear or non-linear in 2 or 3 dimensions.
* **model validation:**  The process of confirming the fidelity of statistical procedures used in the model construction as well as predictive ability of the model.  Does it follow real life randomness, checking the variance vs bias tradeoff, or does the model prediction hold relatively stable and more data is fed to the prediction.
* **accuracy:**  The percentage of correct predictions out of all predictions made by the model.
* **cross-validation:**  Process of further subsetting the training data further in which to train and validate the data multiple times.  In Machine Learning the popular type is k-fold cross validation in which the training set is broken into k subsets and each time a different subset is evaluated based on the training of the remaining subsets.

According to the charts below the performance always tended to plateau for the DecisionTree classifiers but not necessarily for the NaiveBayes classifiers.  Out of the NB classifiers only the GaussianNB classifier plateaued.

```python
from sklearn import naive_bayes, tree, metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

percentages = np.arange(0.05, 0.95, 0.05)

BernoulliNB_accuracy = []
ComplementNB_accuracy = []
GaussianNB_accuracy = []
MultinomialNB_accuracy = []

DecisionTreeClassifier_accuracy = []
DecisionTreeRegressor_accuracy = []
ExtraTreeClassifier_accuracy = []
ExtraTreeRegressor_accuracy = []

for i in percentages:
    X_train, X_test, y_train, y_test = train_test_split(iris.data,
                                                    iris.target,
                                                    train_size=i,
                                                    random_state=0)
    
    classifier = naive_bayes.BernoulliNB()
    classifier.fit(X=X_train, y=y_train)
    predicted = classifier.predict(X_test)
    BernoulliNB_accuracy.append(accuracy_score(y_test, predicted))
   
    classifier = naive_bayes.ComplementNB()
    classifier.fit(X=X_train, y=y_train)
    predicted = classifier.predict(X_test)
    ComplementNB_accuracy.append(accuracy_score(y_test, predicted))
       
    classifier = naive_bayes.GaussianNB()
    classifier.fit(X=X_train, y=y_train)
    predicted = classifier.predict(X_test)
    GaussianNB_accuracy.append(accuracy_score(y_test, predicted))

    classifier = naive_bayes.MultinomialNB()
    classifier.fit(X=X_train, y=y_train)
    predicted = classifier.predict(X_test)
    MultinomialNB_accuracy.append(accuracy_score(y_test, predicted))
    
    ## DTs
    classifier = tree.DecisionTreeClassifier()
    classifier.fit(X=X_train, y=y_train)
    predicted = classifier.predict(X_test)
    DecisionTreeClassifier_accuracy.append(accuracy_score(y_test, predicted))
  
    classifier = tree.DecisionTreeRegressor()
    classifier.fit(X=X_train, y=y_train)
    predicted = classifier.predict(X_test)
    DecisionTreeRegressor_accuracy.append(accuracy_score(y_test, predicted))
    
    classifier = tree.ExtraTreeClassifier(random_state=0)
    classifier.fit(X=X_train, y=y_train)
    predicted = classifier.predict(X_test)
    ExtraTreeClassifier_accuracy.append(accuracy_score(y_test, predicted))
    
    classifier = tree.ExtraTreeRegressor(random_state=0)
    classifier.fit(X=X_train, y=y_train)
    predicted = classifier.predict(X_test)
    ExtraTreeRegressor_accuracy.append(accuracy_score(y_test, predicted))

df = pd.DataFrame(data=np.c_[percentages,
                        BernoulliNB_accuracy,
                        ComplementNB_accuracy,
                        GaussianNB_accuracy,
                        MultinomialNB_accuracy,
                        DecisionTreeClassifier_accuracy,
                        DecisionTreeRegressor_accuracy,
                        ExtraTreeClassifier_accuracy,
                        ExtraTreeRegressor_accuracy],
                  columns=['percentages',
                           'BernoulliNB_accuracy',
                           'ComplementNB_accuracy',
                           'GaussianNB_accuracy',
                           'MultinomialNB_accuracy',
                           'DecisionTreeClassifier_accuracy',
                           'DecisionTreeRegressor_accuracy',
                           'ExtraTreeClassifier_accuracy',
                           'ExtraTreeRegressor_accuracy']
                 )

## There's got to be a better way to get all columns that tie for max for a particular row
## These are best performers for the 50/50 train/test split
print('The best classifiers at 50/50 split are:\n', 
      df.columns[(df.loc[df['percentages'] == 0.5].values==df.loc[df['percentages'] == 0.5].values.max())[0]].values)
print(df.loc[df['percentages'] == 0.5])

# Plot each classifier to see if they plateau from small training sets to large training sets
fig, axs = plt.subplots(2, 4, figsize=(12,8))
xlim = (0, 1)
ylim = (0, 1)
plt.setp(axs, xlim=xlim, ylim=ylim)
fig.suptitle('Increasing training set size NaiveBayes/DecisionTrees')
axs[0,0].plot(percentages, BernoulliNB_accuracy)
axs[0,0].set_title('BernoulliNB')
axs[0,0].axvspan(0.49, 0.51, alpha=0.5, color='yellow')
axs[0,1].plot(percentages, ComplementNB_accuracy)
axs[0,1].set_title('ComplementNB')
axs[0,1].axvspan(0.49, 0.51, alpha=0.5, color='yellow')
axs[0,2].plot(percentages, GaussianNB_accuracy)
axs[0,2].set_title('GaussianNB')
axs[0,2].axvspan(0.49, 0.51, alpha=0.5, color='yellow')
axs[0,3].plot(percentages, MultinomialNB_accuracy)
axs[0,3].set_title('MultinomialNB')
axs[0,3].axvspan(0.49, 0.51, alpha=0.5, color='yellow')
axs[1,0].plot(percentages, DecisionTreeClassifier_accuracy)
axs[1,0].set_title('DecisionTreeClassifier')
axs[1,0].axvspan(0.49, 0.51, alpha=0.5, color='yellow')
axs[1,1].plot(percentages, DecisionTreeRegressor_accuracy)
axs[1,1].set_title('DecisionTreeRegressor')
axs[1,1].axvspan(0.49, 0.51, alpha=0.5, color='yellow')
axs[1,2].plot(percentages, ExtraTreeClassifier_accuracy)
axs[1,2].set_title('ExtraTreeClassifier')
axs[1,2].axvspan(0.49, 0.51, alpha=0.5, color='yellow')
axs[1,3].plot(percentages, ExtraTreeRegressor_accuracy)
axs[1,3].set_title('ExtraTreeRegressor')
axs[1,3].axvspan(0.49, 0.51, alpha=0.5, color='yellow')

for ax in axs.flat:
    ax.set(xlabel='training set size', ylabel='accuracy')

for ax in axs.flat:
    ax.label_outer()
```
